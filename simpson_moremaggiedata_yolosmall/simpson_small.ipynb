```python
# 이미 훈련이 끝난 모델의 confusion matrix 시각화
import os
import cv2
import numpy as np
from ultralytics import YOLO
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. 학습된 YOLO 모델 불러오기
model = YOLO("yolo11n_trained.pt")  # 이미 학습된 모델 가중치 파일 불러오기

# 2. 클래스 이름 정의
class_names = ['bart_simpson', 'homer_simpson', 'lisa_simpson', 'maggie_simpson', 'marge_simpson']

# 3. 테스트 데이터셋 경로 설정
test_images_dir = r'./Test_Simpson_Dataset/dataset/test/images'

# 4. 실제 레이블과 예측 레이블 저장 리스트 생성
true_labels = []
pred_labels = []

# 5. 각 클래스별로 테스트 이미지 예측
for class_name in class_names:
    class_dir = os.path.join(test_images_dir, class_name)
    image_files = [f for f in os.listdir(class_dir) if f.endswith('.jpg') or f.endswith('.png')]

    for image_file in image_files:
        # 이미지 불러오기
        image_path = os.path.join(class_dir, image_file)
        image = cv2.imread(image_path)

        # 모델을 통해 예측 수행
        result = model(image)
        predicted_class = result[0].boxes.cls.cpu().numpy()  # 예측된 클래스

        # 실제 레이블 추가
        true_label = class_names.index(class_name)
        true_labels.append(true_label)

        # 예측된 클래스 레이블 추가 (검출된 첫 번째 객체 기준)
        if len(predicted_class) > 0:
            pred_labels.append(int(predicted_class[0]))
        else:
            pred_labels.append(-1)  # 검출되지 않은 경우

# 6. Confusion Matrix 계산
cm = confusion_matrix(true_labels, pred_labels, labels=[0, 1, 2, 3, 4])

# 7. Confusion Matrix 시각화
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.savefig("confusion_matrix.png")  # 혼동 행렬을 이미지로 저장
plt.show()
```
